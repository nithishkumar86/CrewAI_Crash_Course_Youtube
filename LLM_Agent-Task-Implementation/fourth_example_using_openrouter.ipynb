```python
# research_crew.py
from crewai import Agent, Task, Crew,LLM
from dotenv import load_dotenv
import os
```

```python
load_dotenv()
```

```python
llm = LLM(
    model='openrouter/meta-llama/llama-3.3-70b-instruct:free',
    base_url="https://openrouter.ai/api/v1",
    temperature=0.2
)
```

```python
# Create Data Analyst Agent
data_analyst = Agent(
    role='Data Analyst',
    goal='Analyze data patterns and provide actionable business insights',
    backstory="""You are an experienced data analyst with expertise in 
    statistical analysis and data visualization. You excel at finding 
    meaningful patterns in data and translating them into business value.""",
    verbose=True,
    allow_delegation=False,
    llm=llm
)

# Create Analysis Task
analysis_task = Task(
    description="""Analyze the following sales data and provide insights:

Monthly Sales Data (in thousands):
- January: $120
- February: $135
- March: $142
- April: $138
- May: $155
- June: $168

Provide:
1. Trend analysis
2. Key observations
3. Predictions for next quarter
4. Recommendations for growth""",
    agent=data_analyst,
    expected_output="A comprehensive data analysis report with insights and recommendations stict with 500 tokens"
)
```

```python
# Create Crew
crew = Crew(
    agents=[data_analyst],
    tasks=[analysis_task],
    verbose=True
)
```

```python
# Execute
if __name__ == "__main__":
    result = crew.kickoff()
    print("\n" + "="*50)
    print("ANALYSIS REPORT:")
    print("="*50)
    print(result)
```
