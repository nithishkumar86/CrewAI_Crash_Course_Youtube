# Human Input (optional)	

human_input	Optional[bool]	

Whether the task should have a human review the final answer of the agent. Defaults to False.

```python
import os
from dotenv import load_dotenv
load_dotenv()
```

```python
from crewai import Agent, Task, Crew,LLM
from langchain_community.utilities import GoogleSerperAPIWrapper # THIRD PARTY 
from crewai.tools import tool
```

```python
llm = LLM(
    model="groq/llama-3.3-70b-versatile",
    temperature=0.2,
    api_key=os.getenv("GROQ_API_KEY")
    )
```

```python
@tool()
def google_search(query: str) -> str:
    """perform a web search based on the user query"""
    search = GoogleSerperAPIWrapper()
    return search.run(query)
```

```python
research_agent = Agent(
    role="Senior Web Research Specialist",
    goal="Gather accurate, relevant, and up-to-date information from the web using the google_search tool based on the user's query.",
    backstory=(
        "You are an experienced research analyst who specializes in "
        "collecting high-quality factual information from online sources. "
        "You are skilled at identifying credible information, filtering "
        "irrelevant content, and summarizing findings clearly. "
        "You never fabricate information and strictly rely on real search results."
    ),
    llm = llm,
    tools=[google_search],
    verbose=True,
    max_tokens = 500
)
```

```python
research_task = Task(
    description=(
        "Perform a web search using the google_search tool for the topic: {user_input}. "
        "Collect relevant and recent information from reliable sources. "
        "Summarize key findings and include reference links."
    ),
    expected_output=(
        "A concise research summary containing key findings and relevant "
        "source URLs related to the user's topic."
    ),
    agent=research_agent
)
```

```python
report_writer_agent = Agent(
    role="Professional Report Writer",
    goal="Create a structured, clear, and professional report based strictly on the provided research findings.",
    backstory=(
        "You are a highly experienced report writer specializing in business "
        "and technical documentation. You transform raw research into clear, "
        "structured, and professional reports suitable for decision-makers. "
        "You never introduce information that is not present in the provided context."
    ),
    llm = llm,
    verbose=True,
    max_tokens = 500
)
```

```python
report_task = Task(
    description=(
        "Using the research findings provided, generate a comprehensive "
        "report on the topic: {user_input}. Structure the report clearly "
        "with an introduction, main sections, and conclusion. "
        "Use only the provided research context."
    ),
    expected_output=(
        "A complete professional report including a clear title, introduction, "
        "structured sections, conclusion, and references."
    ),
    human_input = True,
    agent=report_writer_agent
)
```

```python
# Instantiate your crew with a sequential process
crew = Crew(
    agents=[research_agent, report_writer_agent],
    tasks=[research_task, report_task],
    verbose=True
)
```

```python
# Get your crew to work!
result = crew.kickoff(inputs={"user_input":"search recent trends in AI"})

print("######################")
print(result)
```
