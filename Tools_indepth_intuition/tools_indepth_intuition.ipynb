# tools

A tool in CrewAI is a skill or function that agents can utilize to perform various actions.



```python
import os
from dotenv import load_dotenv
load_dotenv()
```

```python
from crewai import Task,Agent,LLM,Crew
from crewai.tools import tool
from crewai_tools import TavilySearchTool # web search extract,search
from crewai_tools import FirecrawlScrapeWebsiteTool # extract the context from specific website
```

```python
#llm Initialization
 
llm = LLM(
    model="groq/llama-3.3-70b-versatile",  #llama-3.1-8b-instant
    temperature=0.2 # for output 
)
```

```python
# Initialize the tool
tavily_tool = TavilySearchTool(max_results = 4,search_depth = 'advanced')
```

```python
# Create an agent with clear instructions
researcher = Agent(
    role='Market Researcher with Specialist in Market Trends',
    goal='Identify the top 3 most impactful AI trends from 2026 based on industry adoption and innovation.',
    backstory='Senior technology analyst with 10 years of experience tracking AI industry developments and emerging technologies.',
    llm = llm,
    tools=[tavily_tool],
    verbose=True, # logs
    max_tokens=500 # to restricct the token limit
)

# Create a task with specific requirements
research_task = Task(
    description="""
    Search for and identify the top 3 AI trends from 2026.
    
    Steps:
    1. Use Tavily to search "top AI trends 2026"
    2. Select 3 trends based on industry impact and adoption
    3. Summarize each trend in 2-3 sentences
    4. Keep total response under 500 tokens
    """,
    expected_output='A concise report listing the top 3 AI trends from 2026, with each trend including its name, a brief description, and why it matters. Maximum 500 tokens total.',
    agent=researcher
)
```

```python
# Form the crew and kick it off
crew = Crew(
    agents=[researcher],
    tasks=[research_task],
    verbose=True 
)
```

```python
result = crew.kickoff()
print(result)
```

# second Example of Inbuilt Tools

```python
import os
from dotenv import load_dotenv
load_dotenv()
```

```python
# Initialize the tool
firecrawl_tool = FirecrawlScrapeWebsiteTool(page_options='onlyMainContent',extractor_options="don't extract url or https extract main content")
```

```python
# Create an agent with clear instructions
content_analyst = Agent(
    role='Website Content Analyst',
    goal='Extract and analyze key information from the given url including features, pricing, and use cases.',
    backstory='Experienced web content analyst specializing in technical documentation and product analysis for developer tools.',
    llm = llm,
    tools=[firecrawl_tool],
    verbose=True,
    max_tokens=500
)

# Create a task with specific requirements
scrape_task = Task(
    description="""
    Scrape and analyze the given url

    url : {url}
    
    Steps:
    1. Use FirecrawlScrapeWebsiteTool to scrape the url
    2. Extract main product features and capabilities
    3. Identify pricing information if available
    4. Note primary use cases or target audience
    5. Keep total response under 500 tokens
    """,
    expected_output='A concise summary of the url covering: what the product does, main features, pricing (if found), and target users. Maximum 500 tokens total.',
    agent=content_analyst
)
```

```python
def main():
    crew = Crew(
        agents = [content_analyst],
        tasks = [scrape_task],
        verbose = True
    )
    result = crew.kickoff(inputs={"url":'https://docs.crewai.com/en/introduction'})
    return result
```

```python
result = main()
print(result)
```

# Custom Tools

```python
from crewai.tools import tool
```

```python
import fitz
import re

@tool(result_as_answer=True)
def pdf_extractor(path:str) -> str:
    """
    Custom tool for extracting content and text from PDF or DOCX files.
    Returns cleaned text content from the document.
    """
    path = r"D:\CREWAI_CRASH_COURSE\Tools_indepth_intuition\resume.pdf"

    text_blocks = []
    with fitz.open(path) as doc:
        for page in doc:
            text = page.get_text("text").strip()
            if text:
                text_blocks.append(text)

    paragraphs = []
    for page_text in text_blocks:
        chunks = re.split(r'\n\s*\n', page_text)
        for chunk in chunks:
            clean = chunk.strip()
            paragraphs.append(clean)
    
    result = "\n\n".join(paragraphs)
    return result
```

```python
resume_parser_agent = Agent(
    role="Resume Parser",
    goal="Extract text content from resume documents using the pdf_extractor tool. "
         "Read the document and return its contents accurately without modification."
         "strictly don't add any other unnescessary content",
    backstory="You are a document processing specialist. Your sole responsibility is to use "
              "the pdf_extractor tool to read documents and extract their text content. "
              "You do not add,modify, summarize, or interpret the content.",
    llm=llm,
    max_iter=3,
    verbose=True,
    tools=[pdf_extractor],
    allow_delegation=False
)

task1_parse_resume = Task(
    description="""
    Use the pdf_extractor tool to load and extract all text content from the resume document.
    
    Instructions:
    - Call the pdf_extractor tool to read the document
    - striclty Extract all text content without any modification
    - Return the complete raw resume text as extracted
    - Don't add unnecessary content to the raw text keep in mind
    Do NOT summarize, interpret, modify or add the extra content by youself to the text.
    keep in your mind, your task is to parse the content not modifiying or adding extra content",
    """,
    expected_output="Complete extracted resume text",
    agent=resume_parser_agent
)
```

```python
def main1():
    crew = Crew(
        agents = [resume_parser_agent],
        tasks = [task1_parse_resume],
        verbose = True
    )
    result = crew.kickoff()
    return result
```

```python
result  = main1()
print(result)
```

# Third Party Tools

```python
from crewai import Agent, Task
from langchain_community.tools.pubmed.tool import PubmedQueryRun #epecifically for medical related
from crewai.tools import tool
```

```python
# CrewAI-native tools (decorated with @tool)

@tool()
def pubmed_search(query: str) -> str:
    """Search PubMed database for medical research articles."""
    pubmed = PubmedQueryRun()
    return pubmed.run(query)
```

```python
# Create an agent with clear instructions
medical_researcher = Agent(
    role='Medical Research Analyst',
    goal='Search PubMed database for the latest peer-reviewed research on specific diseases, treatments, and medical conditions.',
    backstory='Clinical research specialist with 8 years of experience analyzing medical literature and synthesizing findings from peer-reviewed journals.',
    llm=llm,
    tools=[pubmed_search],
    verbose=True,
    max_tokens=500
)

# Create a task with specific requirements
research_task = Task(
    description="""
    User Question :
    {disease}
    
    Steps:
    1. Use the PubMed tool to search for user given disease"
    2. Focus on recent publications and clinical findings
    3. Identify key findings on causes, symptoms, or treatments
    4. Summarize the most relevant research findings
    5. Keep total response under 500 tokens
    """,
    expected_output='A concise medical research summary including: disease overview, recent findings, current treatment approaches, and key takeaways from peer-reviewed sources. Maximum 500 tokens total.',
    agent=medical_researcher
)
```

```python
def main2():
    crew = Crew(
        agents = [medical_researcher],
        tasks = [research_task],
        verbose = True
    )
    result = crew.kickoff(inputs={"disease":"severe fever"})
    return result
```

```python
result = main2()
print(result)
```
