````python
from crewai import Agent, Task, Crew,LLM
import os
from dotenv import load_dotenv
load_dotenv()
````

````python
llm = LLM(
    model = "ollama/gemma3:4b-cloud", # supports multimodel
    base_url = "http://localhost:11434",
    temperature  =0.2
)
````

````python
# Create a multimodal agent
image_analyst = Agent(
    role="Image Analyst",
    goal="Analyze the images and provide detailed descriptions",
    backstory="Expert in visual image analysis with deep knowledge of design and features",
    llm = llm,
    multimodal=True, # this is first time i using this feature
    verbose = True
)

# Create a task for image analysis
task = Task(
    description="Analyze the image black_dog.jpg and provide a detailed description",
    expected_output="A detailed description of the image",
    agent=image_analyst
)
````

````python

# Create and run the crew
crew = Crew(
    agents=[image_analyst],
    tasks=[task],
    verbose = True
)

result = crew.kickoff()
````

````python
print(result.raw,end=" ")
````

````python
from crewai import Agent, Task, Crew, LLM
from pydantic import BaseModel, Field
from typing import List
````

````python
# â”€â”€ Pydantic Models â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€

class FoodItem(BaseModel):
    name: str = Field(description="Name of the food item")
    quantity: str = Field(description="Estimated quantity or portion size")
    calories: float = Field(description="Estimated calories in kcal")
    protein_g: float = Field(description="Protein content in grams")
    carbs_g: float = Field(description="Carbohydrates content in grams")
    fat_g: float = Field(description="Fat content in grams")


class NutritionReport(BaseModel):
    meal_name: str = Field(description="Type of meal eg Breakfast, Lunch, Dinner, Snack")
    food_items: List[FoodItem] = Field(description="List of all food items detected in the image")
    total_calories: float = Field(description="Total calories of the entire meal in kcal")
    total_protein_g: float = Field(description="Total protein of the entire meal in grams")
    total_carbs_g: float = Field(description="Total carbohydrates of the entire meal in grams")
    total_fat_g: float = Field(description="Total fat of the entire meal in grams")
    health_score: int = Field(description="Overall health score of the meal from 1 to 10")
    health_summary: str = Field(description="Brief 2-3 sentence health assessment of the meal")
    suggestions: List[str] = Field(description="List of 3 suggestions to make the meal healthier")
````

````python
# â”€â”€ Agent â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€

nutritionist = Agent(
    role="Senior Certified Nutritionist",
    goal=(
        "Carefully analyze food images to identify every food item, "
        "estimate accurate portion sizes, calculate detailed nutritional content, "
        "and provide actionable health insights"
    ),
    backstory=(
        "You are a certified nutritionist with 15 years of experience in dietary analysis. "
        "You have deep expertise in identifying food items from images, estimating portion sizes, "
        "and calculating macronutrients accurately. You always provide honest, science-based "
        "nutritional assessments and practical health improvement suggestions."
    ),
    llm=llm,
    multimodal=True,
    verbose=True
)

# â”€â”€ Task â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€

nutrition_task = Task(
    description=(
        "Analyze the food image 'food_images.jpg' and perform a complete nutritional analysis.\n\n"
        "Follow these steps carefully:\n"
        "1. Identify every food item visible in the image\n"
        "2. Estimate the portion size of each item\n"
        "3. Calculate calories, protein, carbs, and fat for each item individually\n"
        "4. Sum up the total macronutrients for the full meal\n"
        "5. Give an honest health score from 1 (very unhealthy) to 10 (very healthy)\n"
        "6. Provide a brief health summary and 3 practical suggestions to improve the meal\n\n"
        "Be as accurate as possible with your estimates based on visual portion sizes."
    ),
    expected_output=(
        "A fully structured nutritional report containing: "
        "each food item with its individual macros, "
        "total meal macronutrients, "
        "a health score out of 10, "
        "a health summary, "
        "and 3 suggestions to improve the meal"
    ),
    output_pydantic=NutritionReport,
    agent=nutritionist
)
````

````python
# â”€â”€ Crew â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€

crew = Crew(
    agents=[nutritionist],
    tasks=[nutrition_task],
    verbose=True
)
````

````python
# â”€â”€ Run & Display Results â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€

result = crew.kickoff()
report: NutritionReport = result.pydantic

print("\n" + "="*55)
print("         ðŸ¥— NUTRITIONAL ANALYSIS REPORT")
print("="*55)
print(f"  Meal     : {report.meal_name}")
print(f"  Health   : {'â­' * report.health_score} ({report.health_score}/10)")
print("-"*55)
print(f"  {'FOOD ITEM':<20} {'CAL':>6} {'PRO':>6} {'CARB':>6} {'FAT':>6}")
print(f"  {'':.<20} {'kcal':>6} {'(g)':>6} {'(g)':>6} {'(g)':>6}")
print("-"*55)

for item in report.food_items:
    print(f"  {item.name:<20} {item.calories:>6.0f} {item.protein_g:>6.1f} {item.carbs_g:>6.1f} {item.fat_g:>6.1f}")
    print(f"  {'Portion: ' + item.quantity:<20}")

print("-"*55)
print(f"  {'TOTAL':<20} {report.total_calories:>6.0f} {report.total_protein_g:>6.1f} {report.total_carbs_g:>6.1f} {report.total_fat_g:>6.1f}")
print("="*55)
print(f"\n  ðŸ“‹ Summary:\n  {report.health_summary}")
print("\n  ðŸ’¡ Suggestions to Improve:")
for i, tip in enumerate(report.suggestions, 1):
    print(f"   {i}. {tip}")
print("="*55)
````

````python
import json
import re

def safe_parse_response(response):
    
    # Step 1 - Check if response is empty
    if not response or not response.strip():
        print("âŒ Response is empty")
        return None
    
    # Step 2 - Try direct JSON parse
    try:
        parsed = json.loads(response)
        print("âœ… Parsed directly")
        return parsed
    except json.JSONDecodeError:
        pass

    # Step 3 - Try extracting JSON block from markdown
    # LLMs often wrap JSON in ```json ... ```
    try:
        json_match = re.search(r"```json\s*(.*?)\s*```", response, re.DOTALL)
        if json_match:
            parsed = json.loads(json_match.group(1))
            print("âœ… Extracted from markdown block")
            return parsed
    except json.JSONDecodeError:
        pass

    # Step 4 - Try extracting raw { } block
    try:
        json_match = re.search(r"\{.*\}", response, re.DOTALL)
        if json_match:
            parsed = json.loads(json_match.group())
            print("âœ… Extracted from curly braces")
            return parsed
    except json.JSONDecodeError:
        pass

    # Step 5 - Nothing worked
    print("âŒ Could not parse response as JSON")
    print("Raw response was:", response)
    return None


# â”€â”€ Usage â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€

parsed = safe_parse_response(response)

if parsed:
    print(parsed)
````
