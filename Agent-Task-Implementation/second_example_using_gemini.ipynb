```python
# research_crew.py
from crewai import Agent, Task, Crew,LLM
from dotenv import load_dotenv
import os
```

```python
load_dotenv()
```

```python

llm = LLM(
    model="gemini/gemini-2.5-flash",
    temperature=0.2
)
```

```python


# Create Writer Agent
content_writer = Agent(
    role='Content Writer with specilist in Techincal skills',
    goal='Create engaging and informative blog posts on technical topics',
    backstory="""You are a skilled technical writer who can transform complex 
    concepts into accessible, engaging content. Your writing is clear, 
    informative, and captivating.""",
    verbose=True,
    allow_delegation=False,
    llm=llm
)

# Create Writing Task
writing_task = Task(
    description="""Write a 500-word blog post about the benefits of using 
    open-source LLM models in production environments. 
    Make it engaging and include:
    1. Introduction
    2. Key benefits
    3. Real-world examples
    4. Conclusion""",
    agent=content_writer,
    expected_output="A well-structured 500-word blog post"
)
```

```python
# Create Crew
crew = Crew(
    agents=[content_writer],
    tasks=[writing_task],
    verbose=True
)
```

```python
# Execute
if __name__ == "__main__":
    result = crew.kickoff()
    print("\n" + "="*50)
    print("BLOG POST:")
    print("="*50)
    print(result)
```
